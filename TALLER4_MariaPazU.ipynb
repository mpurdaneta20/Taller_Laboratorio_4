```python
import urllib
import matplotlib as plt
import numpy as np
import random
```

```python
from scipy.interpolate import interp1d
from scipy import stats
%pylab inline
```

```python
urllib.urlretrieve("https://github.com/forero/ComputationalMethodsData/raw/master/homework/hw_5/sparse_sample_monthrg.dat","monthrg.dat")
```

```python
A,B,C,D,S,F = [],[],[],[],[],[] 
N,T=[],[] 

File = open('monthrg.dat','r')
for i, line in enumerate(File):
    if i >450:
        a,b,c,d = line.split(' ', 4)
        A.append(float(a))
        B.append(float(b))
        C.append(float(c))
        D.append(float(d))

NA=np.array(A)
NB=np.array(B)
NC=np.array(C)
ND=np.array(D)

for i in range(len(NA)):
    W=(NA[i]*365.0)
    M=(NB[i]*30.0)
    s=np.sum([M,W])
    N.append(s)
    i=i+1
NN=np.array(N)
for i in range(len(NN)):
    s=np.sum([NN[i],NC[i]])
    T.append(s)
    i=i+1
NT=np.array(T)
print len(NT)
print len(ND)
```
